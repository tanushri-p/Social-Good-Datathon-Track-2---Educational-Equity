```python
# We defined categories based on the available columns to define what it means for 
# a college to be a "good match" for a student based on their background and needs.

# We take into account a holistic view of affordability metrics based on different student backgrounds.
AFFORDABILITY_COLS = [
    'Average Cost of Attendance Out-of-State',
    'Average Cost of Attendance In-District',
    'Average Cost of Attendance In-State',
    'Out-of-State Average Tuition for First-Time, Full-Time Undergraduates',
    'Average In-State Tuition for First-Time, Full-Time Undergraduates',
    'Average Net Price After Grants, 2018-19',
    'Average Net Price After Grants, 2019-20',
    'Average Net Price After Grants, 2020-21',
    'Average Net Price for Low-Income Students, 2018-19',
    'Average Net Price for Low-Income Students, 2019-20',
    'Average Net Price for Low-Income Students, 2020-21'
]

# We define success in a college based on the retention rate and earnings of students after graduation.
SUCCESS_COLS = [
    'First-Time, Full-Time Retention Rate',
    'Percent Full-time, First-time, Pell Grant Recipients Still Enrolled at Same Institution - 8 Years',
    'Median Earnings of Students Working and Not Enrolled 10 Years After Entry',
]

# Often times when students enter college, they lack the knowledge, resources, and confidence to be "successful".
# These are measures of various resources that colleges offer for student support and success.
RESOURCES_COLS = [
    'Student-to-Faculty Ratio',
    'Percent of Full-Time Faculty',
    'Instructional Expenditures Per Student',
    'Instructional Expenses Per FTE'
]

# Students in minority groups might prioritize attending a college where a large portion of the student body population are students
# that look like them, come from similar backgrounds, and share commonalities. Having others similar to you fosters a sense of belonging
# and community on campus, which is important for confidence and success.
DEMOGRAPHICS_COLS = [
    'Percent of Black or African American Undergraduates',
    'Percent of Latino Undergraduates',
    'Percent of Asian Undergraduates',
    'Percent of American Indian or Alaska Native Undergraduates',
    'Percent of Native Hawaiian or Other Pacific Islander Undergraduates',
    'Percent of White Undergraduates',
    'Percent of Women Undergraduates',
    'Percent of Men Undergraduates',
    'Percent of Undergraduates Age 25 to 64',
]
```

```python
# Our dataset contains many columns that are missing over half the data. Here, we filter those columns out.

# Set a threshold for dropping columns with excessive missing values
threshold = 0.5  # 50% missing values

# Drop columns with more than 50% missing values
crv_cleaned = crv.dropna(thresh=int((1-threshold) * len(crv)), axis=1)

# Fill missing values for numerical columns with the median
numerical_cols = crv_cleaned.select_dtypes(include=['float64', 'int64']).columns
crv_cleaned[numerical_cols] = crv_cleaned[numerical_cols].fillna(crv_cleaned[numerical_cols].median())

# Fill missing values for categorical columns with the mode
categorical_cols = crv_cleaned.select_dtypes(include=['object']).columns
for col in categorical_cols:
    crv_cleaned[col].fillna(crv_cleaned[col].mode()[0], inplace=True)

# Display the cleaned DataFrame
crv_cleaned.head()
```

```python
# Our columns contain a wide variety of data that come in different units and scales. For some variables,
# a higher number indicates a better match, while for other columns, a lower number is better. Here, we
# standardize all column data on a 0 to 1 scale.
def normalize(df, col, invert=False):
    if df[col].isna().all():
        return pd.Series([np.nan] * len(df))
        
    series = df[col].fillna(df[col].median())
    norm = (series - series.min()) / (series.max() - series.min())
    return 1 - norm if invert else norm

# This function computes a score for each category we defined. It finds the average of each normalized data point in each column.
def build_category_score(df, columns, invert=False):
    scores = []
    for col in columns:
        if col in df.columns:
            scores.append(normalize(df, col, invert=invert))
    if len(scores) == 0:
        return None
    return sum(scores) / len(scores)
```

```python
# We score demographics differently, since we take into account which identities a student aligns with. Students can select
# multiple categories from race, gender, and whether or not they are a re-entry student. 
def compute_demographic_score(df, user_identities, weights=None):
    # Map identities to column names
    identity_to_column = {
        "Black": "Percent of Black or African American Undergraduates",
        "African American": "Percent of Black or African American Undergraduates",
        "Latino": "Percent of Latino Undergraduates",
        "Hispanic": "Percent of Latino Undergraduates",
        "Asian": "Percent of Asian Undergraduates",
        "American Indian": "Percent of American Indian or Alaska Native Undergraduates",
        "Alaska Native": "Percent of American Indian or Alaska Native Undergraduates",
        "Pacific Islander": "Percent of Native Hawaiian or Other Pacific Islander Undergraduates",
        "White": "Percent of White Undergraduates",
        "Women": "Percent of Women Undergraduates",
        "Men": "Percent of Men Undergraduates",
        "Adult Learner": "Percent of Undergraduates Age 25 to 64",
    }

    # translate user identities to column names
    chosen_cols = [identity_to_column[i] for i in user_identities if i in identity_to_column]

    if len(chosen_cols) == 0:
        return pd.Series([0] * len(df))

    # default equal weights  
    if weights is None:
        weights = [1/len(chosen_cols)] * len(chosen_cols)

    # extract the demographic percentages
    demo_df = df[chosen_cols].copy()

    # fill missing with median values
    demo_df = demo_df.fillna(demo_df.median())

    # normalize each demographic percent to 0–1
    normalized = (demo_df - demo_df.min()) / (demo_df.max() - demo_df.min())

    # compute weighted score
    weighted_score = normalized.mul(weights).sum(axis=1)

    return weighted_score
```

```python
# We compute scores for each category.
df['affordability_score'] = build_category_score(
    df, AFFORDABILITY_COLS, invert=True
)

df['success_score'] = build_category_score(
    df, SUCCESS_COLS, invert=False
)

df['resources_score'] = build_category_score(
    df, RESOURCES_COLS, invert=False
)
```

```python
# Here, students input demographic information, as well as how much they prioritize certain categories (which must add up to 1), and filters for schools
# that align with their preferences.
user_profile = {
    "identities": ["Black", "Latino"],   # demographic identities
    
    "weights": {                         # match score weights
        "affordability": 0.4,
        "success": 0.3,
        "resources": 0.2,
        "demographics": 0.1
    },

    "filters": {                         # college filtering preferences
        "max_price": 15000,
        "state": "CA",
        "institution_type": "Public"
    }
}
```

```python
# We compute affordability, success, resources
df['affordability_score'] = build_category_score(df, AFFORDABILITY_COLS, invert=True)
df['success_score'] = build_category_score(df, SUCCESS_COLS)
df['resources_score'] = build_category_score(df, RESOURCES_COLS)

# Compute demographic match
df['demographic_support_score'] = compute_demographic_score(
    df,
    user_identities=user_profile["identities"]
)
```

```python
# Computing overall match score

w = user_profile["weights"]

df['match_score'] = (
    w["affordability"] * df['affordability_score'] +
    w["success"] * df['success_score'] +
    w["resources"] * df['resources_score'] +
    w["demographics"] * df['demographic_support_score']
)
```

```python
def filter_colleges(df, max_price=None, state=None, institution_type=None):
    filtered = df.copy()
    
    # Filter by affordability
    if max_price is not None:
        if 'Average Net Price After Grants, 2020-21' in df.columns:
            filtered = filtered[
                filtered['Average Net Price After Grants, 2020-21'] <= max_price
            ]

    # Filter by state
    if state is not None and 'State of Institution' in df.columns:
        filtered = filtered[filtered['State of Institution'] == state]

    # Filter by public/private/etc.
    if institution_type is not None and 'Institution Type' in df.columns:
        filtered = filtered[
            filtered['Institution Type'].str.contains(
                institution_type, case=False, na=False
            )
        ]
    
    return filtered
```

```python
def compute_match_scores(df, user_profile):
    """
    1. Applies college filters
    2. Computes affordability/success/resources scores
    3. Computes demographic support score
    4. Computes final match score
    5. Deduplicates colleges so each institution appears once
    """

    # Unpack user inputs
    identities = user_profile.get("identities", [])
    weights = user_profile.get("weights", {})
    filters = user_profile.get("filters", {})

    # ----------- 1. FILTERING -----------
    df_filtered = filter_colleges(
        df,
        max_price=filters.get("max_price"),
        state=filters.get("state"),
        institution_type=filters.get("institution_type")
    )

    # If filters removed all schools, return empty df early
    if df_filtered.empty:
        print("No colleges match the selected filters.")
        return df_filtered

    # ----------- 2. CATEGORY SCORES -----------

    df_filtered['affordability_score'] = build_category_score(
        df_filtered, AFFORDABILITY_COLS, invert=True
    )

    df_filtered['success_score'] = build_category_score(
        df_filtered, SUCCESS_COLS, invert=False
    )

    df_filtered['resources_score'] = build_category_score(
        df_filtered, RESOURCES_COLS, invert=False
    )

    # ----------- 3. DEMOGRAPHIC MATCH SCORE -----------

    df_filtered['demographic_support_score'] = compute_demographic_score(
        df_filtered,
        user_identities=identities
    )

    # ----------- 4. FINAL MATCH SCORE -----------

    w = weights  # shorter alias
    df_filtered['match_score'] = (
        w.get("affordability", 0) * df_filtered['affordability_score'] +
        w.get("success", 0)       * df_filtered['success_score'] +
        w.get("resources", 0)     * df_filtered['resources_score'] +
        w.get("demographics", 0)  * df_filtered['demographic_support_score']
    )

    # ----------- 5. MAKE COLLEGES UNIQUE -----------

    # Determine correct institution name column
    name_col = None
    for candidate in ["Institution Name", "Institution Name_x", "Institution Name_y", "INSTNM"]:
        if candidate in df_filtered.columns:
            name_col = candidate
            break

    if name_col is None:
        raise KeyError("Could not locate institution name column in dataframe.")

    # Sort by highest match score first
    df_filtered = df_filtered.sort_values("match_score", ascending=False)

    # Drop duplicate institutions but keep the best scoring row
    df_filtered = df_filtered.drop_duplicates(subset=name_col, keep="first")

    return df_filtered
```

```python
# We output the top 10 best match colleges for the student
result = compute_match_scores(df, user_profile)

result[['Institution Name_x', 'match_score']].sort_values(by='match_score', ascending=False).head(10)
```
